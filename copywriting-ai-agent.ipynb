````python
class CopywritingState(TypedDict):
    messages: Annotated[list, add_messages]  # ÂØπËØùÂéÜÂè≤
    requirements: dict  # Ëß£ÊûêÂêéÁöÑÁî®Êà∑ÈúÄÊ±Ç
    search_results: List[dict]  # Â∏ÇÂú∫Á†îÁ©∂ÁªìÊûú
    tone_variations: List[dict]  # ËØ≠Ê∞î/È£éÊ†ºÂèò‰Ωì
    variations: List[dict]  # ÊñáÊ°àÂèò‰Ωì
    evaluations: List[dict]  # ËØÑÂàÜÁªìÊûú
    comparison: dict  # ÊØîËæÉÁªìÊûú
    final_output: str  # ÊúÄÁªàËæìÂá∫
    history: List[dict]  # ÂéÜÂè≤ËÆ∞ÂΩï
````

````python
def analyzer(state: CopywritingState) -> Dict[str, Any]:
    """Analyze the user requirements and extract key details"""
    # Get the latest user message
    user_message = state["messages"][-1].content if state["messages"] else ""

    # Use LLM to extract requirement details
    analyzer_prompt = f"""
    As a professional copywriting strategy analyst, please carefully analyze the following user's copywriting requirements and extract all key information:

    User request: {user_message}

    Please output the following information in JSON format (if any information is missing, return an empty string or an empty array for the corresponding field):
    {{
      "project_basics": {{
        "product_service": "", // Name of the product or service
        "industry": "", // Industry field
        "brand_positioning": "" // Brand positioning
      }},
      "audience": {{
        "target_audience": "", // Basic description of the target audience
        "pain_points": "", // Pain points of the target audience
        "decision_factors": "" // Purchase decision factors
      }},
      "content_requirements": {{
        "purpose": "", // Purpose of the copy (e.g., promotion, brand advertisement)
        "tone": "", // Required tone (formal, humorous, professional, etc.)
        "key_points": [], // Key points to be included
        "call_to_action": "" // Expected call to action
      }},
      "technical_specs": {{
        "constraints": {{
          "length": "", // Word count limit
          "format": "", // Format requirements
          "channels": [] // Publishing channels
        }}
      }},
      "preferences": {{
        "examples": [], // Reference examples
        "style_preferences": "", // Style preferences
        "taboos": [] // Content to avoid
      }}
    }}

    Please return the result strictly in JSON format without any additional explanation or text.
    """
    # Invoke LLM to get the analysis result
    analysis_response = llm.invoke(analyzer_prompt)

    # Attempt to parse the JSON response
    try:
        requirements = json.loads(analysis_response.content)
    except json.JSONDecodeError:
        # If parsing fails, try to extract the JSON part
        import re
        json_pattern = r'```json\s*([\s\S]*?)\s*```|{[\s\S]*}'
        match = re.search(json_pattern, analysis_response.content)
        if match:
            try:
                json_content = match.group(1) or match.group(0)
                # Clean any potential non-JSON characters
                json_content = re.sub(r'//.*?$', '', json_content, flags=re.MULTILINE)
                requirements = json.loads(json_content)
            except:
                # Use default structure if extraction fails
                requirements = default_requirements()
        else:
            # Use default structure if no JSON is found
            requirements = default_requirements()

    # Ensure that all required fields exist to prevent subsequent errors
    ensure_structure(requirements)
    return {"requirements": requirements}

def default_requirements():
    """Return the default requirements structure"""
    return {
      "project_basics": {
        "product_service": "NA",
        "industry": "NA",
        "brand_positioning": "NA"
      },
      "audience": {
        "target_audience": "NA",
        "pain_points": "NA",
        "decision_factors": "NA"
      },
      "content_requirements": {
        "purpose": "NA",
        "tone": "NA",
        "key_points": [],
        "call_to_action": "NA"
      },
      "technical_specs": {
        "constraints": {
          "length": "NA",
          "format": "NA",
          "channels": []
        }
      },
      "preferences": {
        "examples": [],
        "style_preferences": "NA",
        "taboos": []
      }
    }

def ensure_structure(requirements):
    """Ensure the requirements have a complete structure"""
    default = default_requirements()

    # Recursively ensure that all keys exist in nested dictionaries
    def ensure_keys(target, reference):
        for key, value in reference.items():
            if key not in target:
                target[key] = value
            elif isinstance(value, dict) and isinstance(target[key], dict):
                ensure_keys(target[key], value)

    ensure_keys(requirements, default)
````

````python
def market_researcher(state: CopywritingState) -> Dict[str, Any]:
    requirements = state.get("requirements", {})
    product = requirements.get("product_service", "")
    audience = requirements.get("target_audience", "")
    purpose = requirements.get("purpose", "")

    # Step 1: Search for similar products.
    similar_product_query = f"Similar products to {product}"
    similar_products_results = search_tool.invoke(similar_product_query)

    # Step 2: Search for competitor analysis and market trends.
    competitor_query = f"{product} competitor analysis market trends"
    competitor_results = search_tool.invoke(competitor_query)

    # Step 3: Search for successful marketing copy examples.
    successful_copy_query = f"{product} {audience} {purpose} successful marketing copy examples case studies conversion"
    successful_copy_results = search_tool.invoke(successful_copy_query)

    # Combine all the search results into one dictionary.
    combined_search_results = {
        "similar_products": similar_products_results,
        "competitor_analysis": competitor_results,
        "successful_copy": successful_copy_results
    }

    # Create a summary prompt for in-depth market analysis.
    summary_prompt = f"""
    As an experienced market research expert, I need to analyze marketing copy information for "{product}" targeted at "{audience}".

    Below are the combined search results:
    {json.dumps(combined_search_results, ensure_ascii=False, indent=2)}

    Please provide a comprehensive market analysis report that includes:
        1. Market Environment Analysis:
           - Current industry trends and major pain points
           - The core needs and purchasing motivations of the target audience
           - The competitive landscape and the marketing copy positioning of key competitors

        2. Audience Insights:
           - Demographic and psychographic characteristics of the target audience
           - Their language habits and expression styles
           - Key emotional triggers and core values that resonate with them
           - Considerations in their purchase decision-making process

        3. Successful Copy Strategies:
           - The structure and format of high-conversion copy
           - Types of effective headlines and openings
           - Persuasion techniques and storytelling methods used in successful cases
           - Common visual elements and presentation formats

        4. Specific Recommendations:
           - 3-5 potential copy themes or directions
           - Key message points tailored for the target audience
           - Recommended call-to-action phrases and closings
           - Suggestions for testing different copy variants

    Please ensure the analysis is in-depth, practical, and based on real data and case studies so it can be directly applied to the copywriting process.

    Return the result in JSON format with a clear and parse-friendly structure.
    """
    try:
        summary_response = llm.invoke(summary_prompt)
    except Exception as e:
        print(f"Error during summary search: {str(e)}")
        # Return default market insights if there is an error
        return {
            "search_results": {
                "market_trends": [
                    "Emphasize personalized experiences for health tech products", 
                    "Data visualization is trending", 
                    "Multi-function integration is mainstream"
                ],
                "audience_preferences": [
                    "Clear and concise information delivery", 
                    "Focus on product practicality", 
                    "Appreciation for community and sharing features"
                ],
                "competitive_strategies": [
                    "Highlight unique product features", 
                    "Endorsements by health experts", 
                    "User story marketing"
                ],
                "success_factors": [
                    "Clear value proposition", 
                    "Empathy with customer pain points", 
                    "Emphasis on tangible results"
                ],
                "suggested_hooks": [
                    "Break free from traditional health monitoring constraints", 
                    "Take control of your health data, take control of your life", 
                    "Provide scientific backing for fitness outcomes"
                ]
            }
        }

    # Attempt to parse the JSON response from the summary.
    try:
        market_insights = json.loads(summary_response.content)
    except json.JSONDecodeError:
        import re
        json_pattern = r'```json\s*([\s\S]*?)\s*```|{[\s\S]*}'
        match = re.search(json_pattern, summary_response.content)
        if match:
            try:
                market_insights = json.loads(match.group(1) or match.group(0))
            except Exception:
                market_insights = {"raw_search_results": combined_search_results}
        else:
            market_insights = {"raw_search_results": combined_search_results}
    return {"search_results": market_insights}
````

````python
def copywriter(state: CopywritingState) -> Dict[str, Any]:
    """Generate multiple copy variants"""
    requirements = state.get("requirements", {})
    market_insights = state.get("search_results", {})

    # Extract and format all non-empty fields from the requirements.
    def extract_non_empty(data, prefix=""):
        result = []
        if isinstance(data, dict):
            for key, value in data.items():
                if isinstance(value, (dict, list)):
                    nested_result = extract_non_empty(value, f"{prefix}{key}: ")
                    if nested_result:
                        result.extend(nested_result)
                elif value and value != "NA":
                    result.append(f"{prefix}{key}: {value}")
        elif isinstance(data, list) and data:
            items = ", ".join([str(item) for item in data if item])
            if items:
                result.append(f"{prefix}{items}")
        return result

    # Extract key details from the requirements.
    product = requirements.get("project_basics", {}).get("product_service", "")
    if product == "Êú™ÊåáÂÆö" or product == "NA":
        product = ""

    req_details = []

    # Project basics
    basics = requirements.get("project_basics", {})
    if basics:
        for key, value in basics.items():
            if value and value != "NA":
                req_details.append(f"- {key}: {value}")

    # Audience information
    audience = requirements.get("audience", {})
    if audience:
        for key, value in audience.items():
            if value and value != "NA":
                req_details.append(f"- {key}: {value}")

    # Content requirements
    content = requirements.get("content_requirements", {})
    if content:
        for key, value in content.items():
            if key == "key_points" and isinstance(value, list) and value:
                req_details.append(f"- key_points: {', '.join(value)}")
            elif value and value != "NA":
                req_details.append(f"- {key}: {value}")

    # Technical constraints
    constraints = requirements.get("technical_specs", {}).get("constraints", {})
    if constraints:
        for key, value in constraints.items():
            if key == "channels" and isinstance(value, list) and value:
                req_details.append(f"- channels: {', '.join(value)}")
            elif value and value != "NA":
                req_details.append(f"- {key}: {value}")

    # Preferences
    prefs = requirements.get("preferences", {})
    if prefs:
        if prefs.get("examples") and isinstance(prefs["examples"], list):
            req_details.append(f"- References: {', '.join(prefs['examples'])}")
        if prefs.get("style_preferences") and prefs["style_preferences"] != "NA":
            req_details.append(f"- Style Preferences: {prefs['style_preferences']}")
        if prefs.get("taboos") and isinstance(prefs["taboos"], list) and prefs["taboos"]:
            req_details.append(f"- Content to Avoid: {', '.join(prefs['taboos'])}")

    # Get previous copy variants from history if available.
    history = state.get("history", [])
    previous_variations = []
    if history:
        for entry in history:
            if "variations" in entry:
                previous_variations.extend(entry["variations"])

    # Construct the prompt for generating copy variants.
    generation_prompt = f"""
    As an award-winning and seasoned copywriting master, you are tasked with creating 10 distinct, attention-grabbing, and high-conversion copy variants for "{product or 'a specified product/service'}".

    ## Requirements Details:
    {chr(10).join(req_details) if req_details else "Please base your creation on the provided market insights and product attributes."}

    ## Market Insights:
    {json.dumps(market_insights, ensure_ascii=False, indent=2) if market_insights else "Please consider the product features and audience needs."}

    {f"## Previously Used Copies (avoid repetition or excessive similarity): {json.dumps(previous_variations, ensure_ascii=False, indent=2)}" if previous_variations else ""}

    ## Creation Requirements:
    Please create 10 entirely distinct copy variants. Each variant should:
    1. Begin with a strong attention-grabbing hook to immediately capture the target audience.
    2. Maintain consistency with the brand positioning and tone.
    3. Seamlessly incorporate all essential selling points without feeling forced.
    4. Accurately appeal to both the emotional and rational needs of the target audience.
    5. Include a clear and compelling call-to-action (CTA).
    6. Utilize a different writing style, technique, or structure from the others.

    ## Recommended Copywriting Style Directions (feel free to select from or create other styles):
    1. Problem-Solution: Begin by addressing the audience's pain points and demonstrate how the product/service resolves them.
    2. Storytelling: Use a brief narrative or scene to showcase the product‚Äôs value.
    3. Data-Driven: Present concrete numbers and facts to establish credibility.
    4. Emotional Appeal: Engage the reader‚Äôs emotions to build resonance.
    5. Scarcity/Urgency: Emphasize limited-time offers or limited availability to prompt immediate action.
    6. Passionate/Exciting: Convey an energetic and vibrant tone with dynamic language and punchy short sentences.
    7. Minimalist/Cool: Employ a clean, restrained style with simple and precise language along with ample whitespace.
    8. Conversational/Lifestyle: Write in a warm, friendly, and informal tone, as if chatting with a friend.
    9. Artistic/Poetic: Use creative, metaphorical language to evoke strong imagery and an aesthetic appeal.
    10. Humorous/Witty: Incorporate playful, clever language and puns to entertain while communicating the message.
    11. Formal/Authoritative: Adopt a structured, dignified tone that builds trust and authority.
    12. Innovative/Experimental: Try non-traditional narrative techniques and unexpected approaches to capture attention.

    ## Return Format:
    Please return a JSON array of 10 variants. Each variant object should contain the following fields:
    - id: Variant number (e.g., 1 to 10)
    - style_name: Name of the copy style (e.g., "Problem-Solution")
    - style_description: A brief description of the style
    - hook: The attention-grabbing hook used in the copy
    - copy: The complete copy content
    - strength: The main advantage of this variant
    - best_for: The most suitable application scenario or channel

    Return only the standard JSON array with no additional explanation.
    """

    generation_response = llm.invoke(generation_prompt)

    # Attempt to parse the JSON response.
    try:
        variations = json.loads(generation_response.content)
    except json.JSONDecodeError:
        import re
        json_pattern = r'```json\s*([\s\S]*?)\s*```|\[[\s\S]*\]'
        match = re.search(json_pattern, generation_response.content)
        if match:
            try:
                variations = json.loads(match.group(1) or match.group(0))
            except Exception:
                variations = []
        else:
            variations = []
    return {"variations": variations}
````

````python
def evaluator(state: CopywritingState) -> Dict[str, Any]:
    """Evaluate the quality of copy variants with added business and technical metrics."""
    requirements = state.get("requirements", {})
    variations = state.get("variations", [])

    if not variations:
        return {"evaluations": []}

    # Extract product and target audience information.
    product = requirements.get("project_basics", {}).get("product_service", "")
    audience = requirements.get("audience", {}).get("target_audience", "")
    purpose = requirements.get("content_requirements", {}).get("purpose", "")

    # Construct the evaluation prompt.
    evaluation_prompt = f"""
    As an expert in copy evaluation with extensive experience in marketing ROI analysis, please provide a comprehensive assessment of the following {len(variations)} copy variants.

    ## Requirements Details
    {json.dumps(requirements, ensure_ascii=False, indent=2)}

    ## Copy Variants to Evaluate
    {json.dumps(variations, ensure_ascii=False, indent=2)}

    ## Scoring Criteria
    For each variant, please score on a scale of 1-10 for each of the following categories:

    ### 1. Creative Performance Metrics
      - Hook Strength: Ability of the opening hook to capture the target audience's attention.
      - Originality: Freshness and uniqueness of the creative concept.
      - Narrative Flow: Smoothness and coherence of the copy's logical flow.
      - Emotional Resonance: Ability to evoke an emotional response from the target audience.
      - Language Expression: Vividness and overall effectiveness of the language used.

    ### 2. Business Effectiveness Metrics
      - Conversion Potential: Likelihood of driving the desired user actions.
      - Brand Consistency: Alignment with the brand's image and positioning.
      - Audience Match: How well the copy aligns with the target audience's needs and preferences.
      - Differentiation: Degree of distinctiveness compared to competitors.
      - Value Proposition Clarity: Clarity in communicating the product/service's core value.

    ### 3. Technical Execution Metrics
      - Key Points Coverage: Extent to which all essential points are covered.
      - SEO Friendliness: Natural integration of keywords and overall potential for search engine optimization.
      - Readability: Ease of reading and comprehension.
      - Channel Fit: Suitability for the intended publishing channels.
      - Specification Compliance: Adherence to word count limits and formatting requirements.

    ## Business Impact Predictions
    For each copy variant, please additionally provide these quantitative predictions:
    
    1. Predicted CTR (Click-Through Rate Percentage): Provide an estimated CTR as a percentage.
    2. Predicted Conversion Rate (Percentage): Provide an estimated conversion rate as a percentage.
    3. Engagement Potential: Likelihood of users engaging with the copy (1-10).
    4. Brand Lift: Potential impact on brand recognition and favorability (1-10).
    5. ROI Potential: Predicted return on investment potential (1-10).

    ## A/B Testing Recommendations
    For each copy variant, provide actionable A/B testing recommendations:
    - Key elements to test.
    - Possible variant directions.
    - Metrics to measure test performance.

    ## Data Collection Recommendations
    Suggest which data points should be collected to verify the actual performance of the copy.
    (Score any quantitative metrics on a 1-10 scale as applicable.)

    Please return your evaluation results as a JSON array, where each evaluation object includes:
    {{
      "id": <corresponding variant ID>,
      "creative_scores": {{
        "hook_strength": <1-10>,
        "originality": <1-10>,
        "narrative_flow": <1-10>,
        "emotional_resonance": <1-10>,
        "language_expression": <1-10>
      }},
      "business_scores": {{
        "conversion_potential": <1-10>,
        "brand_consistency": <1-10>,
        "audience_match": <1-10>,
        "differentiation": <1-10>,
        "value_clarity": <1-10>
      }},
      "technical_scores": {{
        "key_points_coverage": <1-10>,
        "seo_friendliness": <1-10>,
        "readability": <1-10>,
        "channel_fit": <1-10>,
        "spec_compliance": <1-10>
      }},
      "overall_score": <average score>,
      "business_impact": {{
        "predicted_ctr_percentage": "<estimated CTR percentage, e.g. '7.5%'>",
        "predicted_conversion_rate_percentage": "<estimated conversion rate, e.g. '3.2%'>",
        "engagement_potential": <1-10>,
        "brand_lift": <1-10>,
        "roi_potential": <1-10>
      }},
      "strengths": [list of strengths],
      "weaknesses": [list of weaknesses],
      "improvement_suggestions": [list of suggestions],
      "ab_testing": {{
        "elements_to_test": [list of key elements to test],
        "variant_directions": [list of suggested variant directions],
        "measurement_metrics": [list of metrics to measure performance]
      }},
      "data_collection_recommendations": [list of recommended data points to collect]
    }}

    Return only valid JSON without any additional explanation.
    """

    evaluation_response = llm.invoke(evaluation_prompt)

    # Attempt to parse the JSON response.
    try:
        evaluations = json.loads(evaluation_response.content)
    except json.JSONDecodeError:
        import re
        json_pattern = r'```json\s*([\s\S]*?)\s*```|\[[\s\S]*\]'
        match = re.search(json_pattern, evaluation_response.content)
        if match:
            try:
                json_content = match.group(1) or match.group(0)
                evaluations = json.loads(json_content)
            except:
                evaluations = []
        else:
            evaluations = []
    # print("evaluations",evaluations)
    return {"evaluations": evaluations}
````

````python
def formatter(state: CopywritingState) -> Dict[str, Any]:
    """Formats the results into a readable markdown report with enhanced evaluation metrics.
    Also stores every copy variation record along with related evaluation data in a pandas DataFrame
    for a better view, and saves the merged data into a CSV file.
    """
    import pandas as pd

    requirements = state.get("requirements", {})
    market_insights = state.get("search_results", {})
    variations = state.get("variations", [])
    evaluations = state.get("evaluations", [])
    comparison = state.get("comparison", {})

    # Check if comparison is a list; if so, convert it to a dictionary.
    if isinstance(comparison, list):
        comparison = {"Results": comparison}

    # Consolidate results into a markdown output
    md_output = "# Copy Generation & Evaluation Results\n\n"

    # Requirements Summary (using nested structure)
    md_output += "## Requirements Summary\n\n"
    product = requirements.get("project_basics", {}).get("product_service", "Not Specified")
    audience = requirements.get("audience", {}).get("target_audience", "Not Specified")
    purpose = requirements.get("content_requirements", {}).get("purpose", "Not Specified")
    tone = requirements.get("content_requirements", {}).get("tone", "Not Specified")

    md_output += f"- **Product/Service**: {product}\n"
    md_output += f"- **Target Audience**: {audience}\n"
    md_output += f"- **Objective**: {purpose}\n"
    md_output += f"- **Tone**: {tone}\n\n"

    if requirements.get("project_basics", {}).get("industry"):
        md_output += f"- **Industry**: {requirements['project_basics']['industry']}\n"
    if requirements.get("project_basics", {}).get("brand_positioning"):
        md_output += f"- **Brand Positioning**: {requirements['project_basics']['brand_positioning']}\n"

    key_points = requirements.get("content_requirements", {}).get("key_points", [])
    if key_points:
        md_output += "- **Key Points**:\n"
        for point in key_points:
            md_output += f"  - {point}\n"
    md_output += "\n"

    # Market Insights
    md_output += "## Market Insights\n\n"
    if market_insights:
        for key, value in market_insights.items():
            if isinstance(value, list):
                md_output += f"### {key}\n"
                for item in value:
                    md_output += f"- {item}\n"
                md_output += "\n"
            elif isinstance(value, dict):
                md_output += f"### {key}\n"
                for sub_key, sub_value in value.items():
                    md_output += f"- **{sub_key}**: {sub_value}\n"
                md_output += "\n"
            else:
                md_output += f"### {key}\n{value}\n\n"

    # Copy Variants Table (Enhanced)
    md_output += "## Copy Variants\n\n"
    if variations and len(variations) > 0 and "style_name" in variations[0]:
        md_output += "| Variant ID | Style | Style Description | Opening Hook | Best For | Copy Content |\n"
        md_output += "|------------|-------|-------------------|--------------|----------|--------------|\n"
        for v in variations:
            style = v.get('style_name', 'NA')
            style_desc = v.get('style_description', 'NA')
            best_for = v.get('best_for', 'NA')
            md_output += f"| {v.get('id', 'NA')} | {style} | {style_desc} | {v.get('hook', 'NA')} | {best_for} | {v.get('copy', 'NA')} |\n"
    else:
        md_output += "| Variant ID | Style | Opening Hook | Copy Content |\n"
        md_output += "|------------|-------|--------------|--------------|\n"
        for v in variations:
            md_output += f"| {v.get('id', 'NA')} | {v.get('style', 'NA')} | {v.get('hook', 'NA')} | {v.get('copy', 'NA')} |\n"
    md_output += "\n"

    # Scoring Details
    md_output += "## Scoring Details\n\n"
    if evaluations and len(evaluations) > 0 and "creative_scores" in evaluations[0]:
        md_output += "### Overall Scores\n\n"
        md_output += "| Variant ID | Creative Performance | Business Effectiveness | Technical Execution | Overall Score |\n"
        md_output += "|------------|----------------------|------------------------|---------------------|---------------|\n"
        def format_score(score):
            return f"{score:.1f}" if isinstance(score, float) else score
        for e in evaluations:
            var_id = e.get('id', 'NA')
            creative_scores = e.get('creative_scores', {})
            creative_avg = sum(creative_scores.values()) / len(creative_scores) if creative_scores else 'NA'
            business_scores = e.get('business_scores', {})
            business_avg = sum(business_scores.values()) / len(business_scores) if business_scores else 'NA'
            technical_scores = e.get('technical_scores', {})
            technical_avg = sum(technical_scores.values()) / len(technical_scores) if technical_scores else 'NA'
            overall = e.get('overall_score', 'NA')
            md_output += f"| {var_id} | {format_score(creative_avg)} | {format_score(business_avg)} | {format_score(technical_avg)} | {overall} |\n"
        md_output += "\n"

        md_output += "### Detailed Creative Performance Scores\n\n"
        md_output += "| Variant ID | Hook Strength | Originality | Narrative Flow | Emotional Resonance | Language Expression |\n"
        md_output += "|------------|---------------|-------------|----------------|---------------------|---------------------|\n"
        for e in evaluations:
            var_id = e.get('id', 'NA')
            scores = e.get('creative_scores', {})
            md_output += f"| {var_id} | {scores.get('hook_strength', 'NA')} | {scores.get('originality', 'NA')} | {scores.get('narrative_flow', 'NA')} | {scores.get('emotional_resonance', 'N/A')} | {scores.get('language_expression', 'N/A')} |\n"
        md_output += "\n"

        md_output += "### Detailed Business Effectiveness Scores\n\n"
        md_output += "| Variant ID | Conversion Potential | Brand Consistency | Audience Fit | Differentiation | Value Clarity |\n"
        md_output += "|------------|----------------------|-------------------|--------------|-----------------|---------------|\n"
        for e in evaluations:
            var_id = e.get('id', 'NA')
            scores = e.get('business_scores', {})
            md_output += f"| {var_id} | {scores.get('conversion_potential', 'N/A')} | {scores.get('brand_consistency', 'N/A')} | {scores.get('audience_match', 'N/A')} | {scores.get('differentiation', 'N/A')} | {scores.get('value_clarity', 'N/A')} |\n"
        md_output += "\n"

        md_output += "### Detailed Technical Execution Scores\n\n"
        md_output += "| Variant ID | Key Points Coverage | SEO Friendliness | Readability | Channel Fit | Spec Compliance |\n"
        md_output += "|------------|---------------------|------------------|-------------|-------------|-----------------|\n"
        for e in evaluations:
            var_id = e.get('id', 'N/A')
            scores = e.get('technical_scores', {})
            md_output += f"| {var_id} | {scores.get('key_points_coverage', 'N/A')} | {scores.get('seo_friendliness', 'N/A')} | {scores.get('readability', 'N/A')} | {scores.get('channel_fit', 'N/A')} | {scores.get('spec_compliance', 'N/A')} |\n"
    else:
        md_output += "| Variant ID | Hook Score | Attractiveness | Relevance | Persuasiveness | Creativity | Audience Fit | Overall Score |\n"
        md_output += "|------------|------------|----------------|-----------|----------------|------------|--------------|---------------|\n"
        for e in evaluations:
            if isinstance(e, dict):
                var_id = e.get('id', 'N/A')
                hook_score = e.get('hook_score', 'NA')
                attraction = e.get('attraction', 'NA')
                relevance = e.get('relevance', 'NA')
                persuasion = e.get('persuasion', 'NA')
                creativity = e.get('creativity', 'NA')
                target_fit = e.get('target_fit', 'NA')
                total_score = e.get('total_score', 'NA')
                md_output += f"| {var_id} | {hook_score} | {attraction} | {relevance} | {persuasion} | {creativity} | {target_fit} | {total_score} |\n"

    md_output += "\n"

    # Business Impact Predictions (New Section)
    if evaluations and len(evaluations) > 0 and "business_impact" in evaluations[0]:
        md_output += "## Business Impact Predictions\n\n"
        md_output += "| Variant ID | Predicted CTR | Predicted Conversion | Engagement Potential | Brand Lift | ROI Potential |\n"
        md_output += "|------------|---------------|----------------------|----------------------|------------|---------------|\n"
        for e in evaluations:
            var_id = e.get('id', 'N/A')
            impact = e.get('business_impact', {})
            md_output += f"| {var_id} | {impact.get('predicted_ctr_percentage', 'N/A')} | {impact.get('predicted_conversion_rate_percentage', 'N/A')} | {impact.get('engagement_potential', 'N/A')} | {impact.get('brand_lift', 'N/A')} | {impact.get('roi_potential', 'N/A')} |\n"
        md_output += "\n"

    # Pros and Cons Analysis
    md_output += "## Pros and Cons Analysis\n\n"
    for e in evaluations:
        if isinstance(e, dict):
            var_id = e.get('id', 'N/A')
            title = e.get('hook', 'NA')
            md_output += f"### Variant {var_id}: {title}\n\n"
            md_output += "**Strengths:**\n"
            pros = e.get('pros', e.get('strengths', []))
            if isinstance(pros, list):
                for pro in pros:
                    md_output += f"- ‚úÖ {pro}\n"
            elif isinstance(pros, str):
                md_output += f"- ‚úÖ {pros}\n"

            md_output += "\n**Weaknesses:**\n"
            cons = e.get('cons', e.get('weaknesses', []))
            if isinstance(cons, list):
                for con in cons:
                    md_output += f"- ‚ùå {con}\n"
            elif isinstance(cons, str):
                md_output += f"- ‚ùå {cons}\n"

            suggestions = e.get('improvement_suggestions', '')
            if suggestions:
                md_output += "\n**Improvement Suggestions:**\n"
                if isinstance(suggestions, list):
                    for suggestion in suggestions:
                        md_output += f"- üí° {suggestion}\n"
                elif isinstance(suggestions, str):
                    md_output += f"- üí° {suggestions}\n"

            # A/B Testing Recommendations (New Section)
            ab_testing = e.get('ab_testing', {})
            if ab_testing:
                md_output += "\n**A/B Testing Recommendations:**\n"
                if 'elements_to_test' in ab_testing:
                    md_output += "Suggested Elements to Test:\n"
                    elements = ab_testing['elements_to_test']
                    if isinstance(elements, list):
                        for element in elements:
                            md_output += f"- üîç {element}\n"
                    else:
                        md_output += f"- üîç {elements}\n"
                if 'variant_directions' in ab_testing:
                    md_output += "\nVariant Directions:\n"
                    directions = ab_testing['variant_directions']
                    if isinstance(directions, list):
                        for direction in directions:
                            md_output += f"- üîÑ {direction}\n"
                    else:
                        md_output += f"- üîÑ {directions}\n"
                if 'measurement_metrics' in ab_testing:
                    md_output += "\nMeasurement Metrics:\n"
                    metrics = ab_testing['measurement_metrics']
                    if isinstance(metrics, list):
                        for metric in metrics:
                            md_output += f"- üìä {metric}\n"
                    else:
                        md_output += f"- üìä {metrics}\n"

            # Data Collection Recommendations (New Section)
            data_collection = e.get('data_collection_recommendations', [])
            if data_collection:
                md_output += "\n**Data Collection Recommendations:**\n"
                if isinstance(data_collection, list):
                    for rec in data_collection:
                        md_output += f"- üìà {rec}\n"
                else:
                    md_output += f"- üìà {data_collection}\n"

            md_output += "\n"

    # Copy Comparison
    md_output += "## Copy Comparison\n\n"
    if comparison:
        # Ranking
        ranking = comparison.get('ranking', [])
        if ranking:
            md_output += "### Ranking\n"
            if isinstance(ranking, list):
                for rank_item in ranking:
                    md_output += f"- {rank_item}\n"
            elif isinstance(ranking, dict):
                for rank, item in ranking.items():
                    md_output += f"- **{rank}**: {item}\n"
            elif isinstance(ranking, str):
                md_output += f"{ranking}\n"
            md_output += "\n"
        # Best Hook
        best_hook = comparison.get('best_hook', '')
        if best_hook:
            md_output += f"### Best Hook\n{best_hook}\n\n"
        # Advantages Comparison
        advantages = comparison.get('advantages', {})
        if advantages:
            md_output += "### Advantages Comparison\n"
            if isinstance(advantages, dict):
                for var, adv in advantages.items():
                    md_output += f"- **{var}**: {adv}\n"
            elif isinstance(advantages, list):
                for adv in advantages:
                    md_output += f"- {adv}\n"
            elif isinstance(advantages, str):
                md_output += f"{advantages}\n"
            md_output += "\n"
        # Suitable Scenarios
        scenarios = comparison.get('scenarios', {})
        if scenarios:
            md_output += "### Suitable Scenarios\n"
            if isinstance(scenarios, dict):
                for var, scenario in scenarios.items():
                    md_output += f"- **{var}**: {scenario}\n"
            elif isinstance(scenarios, list):
                for scenario in scenarios:
                    md_output += f"- {scenario}\n"
            elif isinstance(scenarios, str):
                md_output += f"{scenarios}\n"
            md_output += "\n"
        # Recommended Variant
        recommendation = comparison.get('recommendation', '')
        if recommendation:
            md_output += f"### Recommended Variant\n{recommendation}\n\n"
        # Combination Suggestions
        combination = comparison.get('combination', '')
        if combination:
            md_output += f"### Combination Suggestions\n{combination}\n\n"

    # Store every variation record and related evaluation data to a pandas DataFrame.
    # Create a DataFrame for variations.
    df_variations = pd.DataFrame(variations)
    # Create a DataFrame for evaluations (if available).
    df_evaluations = pd.DataFrame(evaluations) if evaluations else pd.DataFrame()
    # Merge the two DataFrames on 'id' if possible.
    if not df_variations.empty and not df_evaluations.empty and 'id' in df_variations.columns and 'id' in df_evaluations.columns:
        df_merged = pd.merge(df_variations, df_evaluations, on='id', suffixes=('_copy', '_eval'))
    else:
        df_merged = df_variations.copy()
    # Save the merged DataFrame to a CSV file.
    # csv_filename = "copy_variations.csv"
    # df_merged.to_csv(csv_filename, index=False)
    
    # Optionally, you could display the DataFrame in your environment:
    # python.display_dataframe_to_user("Copy Variations", df_merged)
    # print(df_merged)

    return {"final_output": md_output, "df": df_merged}
````

````python
def memory_keeper(state: CopywritingState) -> Dict[str, Any]:
    """Store the results of the current session into the history log."""
    current_entry = {
        "requirements": state.get("requirements", {}),
        "search_results": state.get("search_results", {}),
        "variations": state.get("variations", []),
        "evaluations": state.get("evaluations", []),
        "comparison": state.get("comparison", {}),
        "timestamp": pd.Timestamp.now().isoformat()
    }

    # Retrieve the existing history, if available
    history = state.get("history", [])
    history.append(current_entry)

    # To prevent the state from growing too large, only keep the last 5 records
    if len(history) > 5:
        history = history[-5:]

    # Add the final message to the conversation history
    final_message = AIMessage(content=f"I have generated your copy and evaluated it. Here is a summary of the results:\n\n"
                                      f"- Generated {len(state.get('variations', []))} copy variations\n"
                                      f"- Evaluated on 6 dimensions\n"
                                      f"- Provided detailed comparisons and suggestions\n\n"
                                      f"You can view the full report for more details.")

    return {"history": history, "messages": [final_message]}
````
